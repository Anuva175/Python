SECTION B — JSON

Exercise 5

Write only “Delivered” orders to JSON.
Output:
delivered_orders_json/

```python
df_delivered=df.filter(df.order_status=="Delivered")
df_delivered.write.mode("overwrite").json("delivered_order_json")
df_delivered.show()
```

Exercise 6

Read JSON and filter:
city = "Mumbai"
payment_mode = "Card"

```python
df_json=spark.read.json("delivered_order_json")
df_filtered=df_json.filter(
    (df_json.city=="Mumbai")&
    (df_json.payment_mode=="Card")
)
df_filtered.show()
```


Exercise 7

Add a column:
delivery_category
Logic:
delivery_time_minutes > 45 → "Late"
else → "OnTime"
Write output to JSON.


```python
from pyspark.sql.functions import when
df_category=df_filtered.withColumn("delivery_category",when(df_filtered.delivery_time_minutes>45,"Late").otherwise("OnTime"))
df_category.write.mode("overwrite").json("categorized_orders_json")
df_category.show()


```

Exercise 8

Force JSON output to a single partition and observe number of files created.

```python
df_categories=df
df_categories.coalesce(1)\
    .write.mode("overwrite")\
    .json("single_file_orders_json")
df_categories.show()
```

SECTION C — PARQUET

Exercise 9

Convert full dataset to Parquet.
Output:
orders_parquet/

```python
df=spark.read.option("header",True).option("inferScema",True).csv("orders.csv")
df.write.mode("overwrite").parquet("orders_parquet")
df.show()
```

Exercise 10

Read Parquet and filter:
cuisine = "Indian"
order_amount > 500

```python
parquet_df=spark.read.parquet("orders_parquet")
filtered_df=parquet_df.filter(
    (parquet_df.cuisine=="Indian")&
    (parquet_df.order_amount>500)
)
filtered_df.show()
```

Exercise 11

Sort Parquet data by order_amount descending and write top 10 orders back to Parquet.

```python
top10_df=parquet_df.orderBy("order_amount",ascending=False).limit(10)
top10_df.write.mode("overwrite").parquet("top10_orders_parquet")
top10_df.show()
```
